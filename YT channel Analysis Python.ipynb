```python
#Clean the "Rank Column"(removing the ',' 'th', 'st', 'nd', 'rd',) and convert it into int or float

YTdata['Rank']=YTdata['Rank'].str[0:-2]
```

```python
YTdata['Rank']=YTdata['Rank'].str.replace(',','')
```

```python
YTdata['Rank']=YTdata['Rank'].str.replace(',','').astype('int')
```

```python
#Convert the datatype of the "Video Uploads & Subscribers" columns

YTdata['Video Uploads']=YTdata['Video Uploads'].astype('int')
```

```python
YTdata['Subscribers']=YTdata['Subscribers'].astype('int')
```

```python
YTdata.dtypes
```

```python
#Clean the "Grade" column by giving each unique grades numerical attribute & convert datatype to int

YTdata['Grade'].unique()
```

```python
YTdata['Grade']=YTdata['Grade'].map({'A++ ':10, 'A+ ':8, 'A ':6, 'A- ':4, 'B+ ':2})
```
